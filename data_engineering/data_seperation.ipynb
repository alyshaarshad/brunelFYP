```python
import os
from sklearn.preprocessing import normalize
import numpy as np
import matplotlib.pyplot as plt 
import re
import os
import base64




directory = 'new_data'
max_sect_1 = 8
max_sect_2 = 7
max_sect_3 = 5

section_1 = "section 1"
section_2 = "section 2"
section_3 = "section 3"

u_id = 134578
```

```python
def jaw_clench_lines(clean_data):
    
    lines = []
    data = clean_data.split('\n') #split string into a list
    for i,temp in enumerate(data):
        if 'jaw_clench' in temp:
            #print(i)
            lines.append(i)
            
    
            
            
    return lines, data        
```

```python
final_array=[]


for subdir, dirs, files in os.walk(directory):
    for file in files:
        filepath = subdir + os.sep + file
        
        print(filepath)
        
        u_id = u_id+1
    
        if file.endswith(".txt"):
            section = int(re.search('\d+', filepath).group(0))
        
            slides = []
            first_clench = []
            second_clench = []


       
            with open(filepath, 'r') as f:
                data = f.read()
                

                lines, data_list = jaw_clench_lines(data)
                
                data_list = [x for x in data_list if "jaw_clench" not in x]


            
                if section_3 not in filepath:
                
                    slides = lines[-7:]
                    N=5

                else:
                    slides = lines[-5:]
                    N=6
       


        #print('potential slide jc = ' , slides)   
        #print(' difference between slide jc =', np.diff(slides))

            l = [i for i in range(len(np.diff(slides)))]



            for i in l:
                first_clench.append(slides[i])
                second_clench.append(slides[i+1])


            zipped_range = list(zip(first_clench,second_clench))
        
            print(filepath)
            #print(zipped_range)
            #print(len(zipped_range))
            #print()
    
        
     
        
            for i,each in enumerate(zipped_range):
                start = each[0]
                end = each[1]

           

                list_data = data_list[start:end]
            

                subList = [list_data[n:n+N] for n in range(0, len(list_data), N)]
                
                if section_2 not in filepath:
                    slide_number = i+1
                    print(slide_number)
                    
                else:
                    slide_number = 7+i
                    print(slide_number)

                for row in subList:
                    h = ''.join(map(str, row))
                    e = h.replace("-∞", "0.00")
                 
                    k = re.findall("\d+\.\d+", e)
    
                    q = [u_id,section,slide_number]
    
                    k.extend(q)

                    final_array.append(k)
                
                

        #
   
        
            
```
