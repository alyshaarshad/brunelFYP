    ii.if less than 100 the remove data inbetween the two points combining the jaw clench - done
3. seperate the data and collect the data from the same section

```python
import os
from sklearn.preprocessing import normalize
import numpy as np
import matplotlib.pyplot as plt 

directory = 'data'

max_sect_1 = 8
max_sect_2 = 7
max_sect_3 = 5

section_1 = "section 1"
section_2 = "section 2"
section_3 = "section 3"

data_included = ["/muse/elements/alpha_absolute" , 
               "/muse/elements/beta_absolute",
               "/muse/elements/delta_absolute",
               "/muse/elements/theta_absolute",
               "/muse/elements/gamma_absolute",
               "/muse/elements/jaw_clench"]

data_included_3 = ["/muse/acc","/muse/gyro", "/muse/elements/jaw_clench"]

```

```python
for subdir, dirs, files in os.walk(directory):
    for file in files:
        filepath = subdir + os.sep + file
        
        
        
        
        if file.endswith(".txt"):
            jaw_count,clean_data = cleandata(filepath)
            
            print()
            print(filepath, jaw_count)
            
            
            
            
            
            lines,data = jaw_clench_lines(clean_data)
            distance = np.diff(lines)

            print("Lines which jaw clenches appear:", lines)
            
            
            
            

            
            y = [i for i in range(len(distance))]
            zipped_distance  = list(zip(y, distance))
            
            print("all distances:", distance)
            
            
            
            
            if (section_1 in filepath and len(zipped_distance)>max_sect_1):
                
                print("yes - section 1 ")
                
                final_removal_jawclench(zipped_distance,lines,data, filepath ,max_sect_1, data_included )
                
            elif (section_2 in filepath and len(zipped_distance)>max_sect_2):
                
                print("yes - section 2")
                
                final_removal_jawclench(zipped_distance,lines,data, filepath , max_sect_2, data_included)
                
            elif (section_3 in filepath and len(zipped_distance)>max_sect_3):
                
                print("yes - section 3")
                
                final_removal_jawclench(zipped_distance,lines,data, filepath, max_sect_3, data_included_3)
                
            else: 
                
                if (section_3 not in filepath):
                    useless_data_removal(data,lines,filepath, data_included)
                
                else:
                    useless_data_removal(data,lines, filepath, data_included_3)
                    
                    
                print()
                print("Distance between each jaw clench: ", zipped_distance)
                
                
            
            plt.scatter(y, distance)
            plt.show()
            
            
        
            
            
            
            
            
```

```python
def cleandata(filepath):
    with open(filepath, 'r') as f2:
                raw_data = f2.read()
                clean_data = raw_data.replace("[8000]", "\n")
                count = clean_data.count('jaw_clench')

    return count,clean_data
```

```python
def jaw_clench_lines(clean_data):
    
    lines = []
    data = clean_data.split('\n') #split string into a list
    for i,temp in enumerate(data):
        if 'jaw_clench' in temp:
            #print(i)
            lines.append(i)
            
    
            
            
    return lines, data
```

```python
def takeSecond(elem):
    return elem[1]

def takefirst(elem):
    return elem[0]
```

```python
def sort_jawclenches(zipped_distance, max_jc):
        zipped_distance.sort(key=takeSecond)
        h = zipped_distance[:-(max_jc)]  
        
        h.sort(key = takefirst)
        
       
        
        
    
    
        return h 
    
```

```python
def removal(ejc, lines,data):
    for i,x in ejc: #iterate through the list of sorted distance shortest and their indexes
        j= i+1
        
        first_clench = lines[i] #identify the index of the first clench in the original list of jaw clenches
        second_clench = lines[j]#identify the index of the second clench in the original list of jaw clenches
        
        for i in range(first_clench, second_clench): #iterate between the two indexes of the jaw clenches and revalue each element
            data[i] = 'remove'
      
             
    return data
        
        
        
        
    
```

```python
def final_removal_jawclench(zipped_distance,lines,data, filepath , max_jc , data_inc ):
    
    
    ejc= sort_jawclenches(zipped_distance , max_jc)
    
    print("there were extra - Distance between each jaw clench: ",ejc)
    new_data = [removal(ejc,lines, data)]
                
               
    k = [x for x in new_data[0] if x != 'remove']
    
    useless_data_removal(k, lines , filepath,  data_inc)
    
    
    
    
        
```

# Cleaning data

```python
def useless_data_removal(data,lines, filepath, data_inc):
    
    data_1 =[]
    
    del data[:lines[0]]
    
    for i , each in enumerate(data):
        for y in data_inc:
            if y in each:
                data_1.append(each) 
            
            
    gh = "\n".join([str(elem) for elem in data_1])
    
    #print(gh)
        
    with open("new_data/"+filepath, "w") as save_data:
        save_data.write(gh) 
    
    
```
